### 2.2 Data Tidiness

#### Issue #1: Concatenate columns and Remove redundant columns

- The four columns that indicate a developer's type, or just their job title, are DeveloperType, WebDeveloperType, MobileDeveloperType, and NonDeveloperType. It is possible for a developer to hold various job titles.
- However, it is not necessary to enlarge to four columns in order to show a single job title. These columns can be combined into a single column. Simply comma-separate the job title if someone has multiple titles.

```python
# Eliminate all NaN values (should the developer have a single job title with NaN values for the others).
# Next, eliminate all non-alphabet characters from the result using regex \W.
clean_df['Job Title'] = clean_df[['DeveloperType', 'WebDeveloperType', 'MobileDeveloperType', 'NonDeveloperType']].apply(lambda x: ''.join(str(x.dropna().values)), axis=1).str.replace(r"\W", ' ', regex=True)
```

```python
# Drop old columns
clean_df.drop(['DeveloperType', 'WebDeveloperType', 'MobileDeveloperType', 'NonDeveloperType'], axis=1, inplace=True)
```

```python
# Show data
clean_df.head()
```

#### Issue #2: Fill in NaN values

- We discovered that several columns had NaN values that needed to be filled in when we looked up the statistics for each column.
- In numerical column values, we shall enter zero (0).
- For categorical columns, we will fill in string `Unknown`

```python
# Collect numeric columns name
colnames_numerics_only = clean_df.select_dtypes(include=np.number).columns.tolist()
colnames_numerics_only
```

```python
# Fill NaN for each column
for col_name in colnames_numerics_only:
  clean_df[col_name].fillna(0, inplace=True)
```

```python
# Show data
clean_df[colnames_numerics_only]
```

```python
# Collect categorical columns name
colnames_categorical_only = clean_df.select_dtypes(include=object).columns.tolist()
print("Total categorical columns: ", len(colnames_categorical_only))
```

```python
# Fill NaN for each column
for col_name in colnames_categorical_only:
  clean_df[col_name].fillna('Unknown', inplace=True)
```

```python
# Show data
clean_df[colnames_categorical_only].head()
```

# Data Visualization

```python
def draw_chart(plt_name, df_name, x_axis, y_axis, rotation, bins, color, size_x, size_y, title, font_size):
  """
  Return histogram of given parameters
  :input: necessary parrameters
  :output: histogramm
  """
  plt_name.figure(figsize=(size_x,size_y))
  plt_name.hist(data = df_name, x = x_axis, bins = bins, color = color)
  plt_name.xticks(rotation = rotation)
  plt_name.title(title, fontsize = font_size)
  plt_name.xlabel(x_axis, fontsize = font_size)
  plt_name.ylabel(y_axis, fontsize = font_size)
  return plt_name
```

### Question 1: Which country has the largest survey respondent?

- This graph will show us which nation has received the most responses, and which has received the fewest.

```python
draw_chart(plt, clean_df[:1000], 'Country', 'Frequency', 90, 200, 'purple', 30, 10, 'Distribution of Country', 14)
```

- As we can see, the United States has the highest percentage of responders, while Paraguay has the lowest. 
- The dataset is rather enormous, so I just used 1000 entries.


### Question 2: Which professional has the largest number of respond of this survey? 

- This graph will show us which occupation has the most respondents overall and which has the fewest.

```python
draw_chart(plt, clean_df, 'Professional', 'Frequency', 0, 80, 'red', 25, 8, 'Distribution of Professional Occupation', 14)
```

- It is evident that the majority of respondents identify as professional developers, while the least number say they had previously worked as professional developers.

### Question 3: What is the average years of experience of all developer?

- This chart will let us know the average years of experience of all developer

```python
draw_chart(plt, clean_df[:1000], 'YearsProgram', 'Frequency', 45, 100, 'green', 30, 8, 'Distribution of Years of Exeperience', 14)
```

- We can see the average years of experience of developers that took part in the survey is 20 or more years


### Question 4: Is there any relation between Education and Salary?

This boxplot will help us find out that any relation between the Education Rank and Salary of the developer.

```python
plt.figure(figsize=(10,8))
base_color = sns.color_palette()[4]
sns.boxplot(data = clean_df[:500], x = 'FormalEducation', y = 'Salary', color = base_color)
plt.title('Salary Distribution by Education Rank', fontsize = 14)
plt.xlabel('Education Rank',fontsize = 14)
plt.ylabel('Salary', fontsize = 14)
plt.xticks(rotation=90);
```

- Doctoral degree has the highest maximum salary, despite of having many outliers
